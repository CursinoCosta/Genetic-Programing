```python
def div(a,b):
    if(b==0):
        return 1
    else:
        return a/b
```

```python
def derivacao(nterm, num):
    if(nterm == '<start>'):
        if(num%10 == 0):
            return ( '<exp> <op> <exp>')
        if(num%10 == 1):
            return ( '<exp> <op> <exp>')
        if(num%10 == 2):
            return ('<exp> <op> <exp> <op> <exp>')
        if(num%10 == 3):
            return ('<exp> <op> <exp> <op> <exp> <op> <exp>')
        if(num%10 == 4):
            return ( '<exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp>')
        if(num%10 == 5):
            return ( '<exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp>')
        if(num%10 == 6):
            return ( '<exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp>')
        if(num%10 == 7):
            return ( '<exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp>')
        if(num%10 == 8):
            return ('<exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp>')
        if(num%10 == 9):
            return ('<exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp> <op> <exp>')
    
    if(nterm == '<exp>'):
        if(num%3 == 0):
            return ('<term> <op> <term>')
        elif(num%3 == 1):
            return ('( <term> <op> <term> )')
        elif(num%3 == 2):
            return ( '<protop> ( <term> , <term> )')
            
    elif(nterm == '<op>'):
        if(num%3 == 0):
            return ( '+')
        elif(num%3 == 1):
            return ('-')
        elif(num%3 == 2):
            return ('*')
        elif(num%3 == 3):
            return ('<protop>')
            
    if(nterm == '<protop>'):
        return ('div')

    if(nterm == '<term>'):
        return (f'x{num%8}')
        
    else:
        return ('t')
```

```python
def dividir_binario(numero, tamanho_segmento=5):
    binario = bin(numero)[2:]
    binario = binario.zfill(((len(binario) + tamanho_segmento - 1) // tamanho_segmento) * tamanho_segmento)
    
    segmentos = [binario[i:i + tamanho_segmento] for i in range(0, len(binario), tamanho_segmento)]
    return [int(segmento, 2) for segmento in segmentos]

i = 0

def incrementar():
    global i
    i += 1   
    
def reset():
    global i
    i = 0

def check():
    return i
    
```

```python
def ConstructTree(nums,no = '<start>', pai=None):
    if(check() >= len(nums)):
        reset()
    no = Node(no, parent=pai)
    
    next = derivacao(no.name, nums[i])
    incrementar()
    if(next == 't'):
        return no
    for s in next.split():
        ConstructTree(nums,s,no)
        
    return no
        
        

def ImprimeArvore(raiz):
    for pre, _, node in RenderTree(raiz):
        print(f"{pre}{node.name}")

def armazena_folhas(raiz):
    folhas = []
    for pre, _, n in RenderTree(raiz):
        if not n.children:
            folhas.append(n.name)  
    return folhas
```

```python
nums = dividir_binario(16549846546846541654465431232323567569749721554964964246126)
print(nums)
arvore = ConstructTree(nums)
folhas = armazena_folhas(arvore)
' '.join(folhas)
```
